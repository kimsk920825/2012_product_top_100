```python
import requests, openpyxl
from bs4 import BeautifulSoup 
```

```python
work_book = openpyxl.Workbook()
work_sheet = work_book.active
work_sheet.title = '지마켓_탑_100'
work_sheet.column_dimensions['A'].width = 10
work_sheet.column_dimensions['B'].width = 60
work_sheet.column_dimensions['C'].width = 15
work_sheet.append(['순위','상품명','가격'])
product_info = list()
```

```python
res = requests.get('http://corners.gmarket.co.kr/Bestsellers?viewType=G&groupCode=G07',headers={'User-Agent':'Mozilla/5.0'}) 
soup = BeautifulSoup(res.content,'html.parser' )
```

```python
best_lists = soup.select('div.best-list')
best_list = best_lists[1]
card = best_list.select('ul>li')
```

```python
for index, item in enumerate(card):
    product = item.select_one('a.itemname')
    price = item.select_one('div.s-price > strong')
    print(index+1, product.get_text(), price.get_text())
    work_sheet.append([index+1, product.get_text(), price.get_text()])
work_book.save('g_market_top_100.xlsx')
work_book.close()
```
