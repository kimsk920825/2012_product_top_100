```python
import requests, openpyxl, re
from bs4 import BeautifulSoup 
from bs4.element import NavigableString
```

```python
work_book = openpyxl.Workbook()
work_sheet = work_book.active
work_sheet.title = '11번가_탑_100'
work_sheet.column_dimensions['A'].width = 10
work_sheet.column_dimensions['B'].width = 60
work_sheet.column_dimensions['C'].width = 15
work_sheet.append(['순위','상품명','가격'])
```

```python
res = requests.get('http://www.11st.co.kr/browsing/BestSeller.tmall?method=getBestSellerMain&cornerNo=5#pageNum%%5',headers={'User-Agent':'Mozilla/5.0'})
```

```python
soup = BeautifulSoup(res.content,'html.parser')
```

```python
data = soup.select('div#bestPrdList div.viewtype')
#data_2 = data.select('div.viewtype')
best_list = data[0]
card = best_list.select('ul>li')
print(card)
```

```python
for index, item in enumerate(card):
    product = item.select_one('div.pname > p')
    price = item.select_one('strong.sale_price')
    print(index+1, product.get_text().strip(''), price.get_text().strip(''))
    work_sheet.append([index+1, product.get_text().strip(''), price])
work_book.save('11번가_top_100.xlsx')
work_book.close()

```
